# **4.3.1.8 LAB: Day of the year: writing and using your own functions**

## **Scenario:**
Your task is to write and test a function which takes three arguments (a year, a month, and a day of the month) and returns the corresponding day of the year, or returns `None` if any of the arguments is invalid.

Use the previously written and tested functions. Add some test cases to the code. This test is only a beginning.

````python
def is_year_leap(year):
    if year % 4 == 0:
        if year % 100 != 0:
            return True
        elif year % 400 == 0:
            return True
        else:
            return False
    else:
        return False

def days_in_month(year, month):
    month_length = [28, 29, 30, 31]
    months_with_30_days = [4, 6, 9, 11]
    months_with_31_days = [1, 3, 5, 7, 8, 10, 12]
    FEBRUARY = 2
    
    if (year < 1582) or (month <= 0) or (month > 12):
        return None
    elif (is_year_leap(year) == False) and (month == FEBRUARY):
        return month_length[0]
    elif (is_year_leap(year) == True) and (month == FEBRUARY):
        return month_length[1]
    elif month in months_with_30_days:
        return month_length[2]
    elif month in months_with_31_days:
        return month_length[3]

def day_of_year(year, month, day):
    if (year < 1582) or (month <= 0) or (month > 12) or (day <= 0) or (day > 31):
        return None
    else:
        corresponding_day = 0
        for rest_of_the_month in range(1, month):
            month_length = days_in_month(year, rest_of_the_month)
            corresponding_day += month_length
        day_in_month = days_in_month(year, month)
        if day >= 1 and day <= day_in_month:
            return corresponding_day + day

test_years = [1920, 2001, 2016, 1987]
test_months = [2, 2, 1, 11]
test_days = [29, 28, 31, 30]
test_results = [60, 59, 31, 334]
for i in range(len(test_years)):
    yr = test_years[i]
    mo = test_months[i]
    dy = test_days[i]
    print(yr, mo, dy, "->", end="")
    result = day_of_year(yr, mo, dy)
    if result == test_results[i]:
        print("OK")
    else:
        print("Failed")

print(day_of_year(2000, 12, 31))
````
